```python
#TIC-TAC-TOE

import math

# Symboles pour les joueurs
ai = 'X'
human = 'O'

# Initialisation des listes de scores et du plateau
score_ai = []
score_human = []
magic_board = [2, 7, 6, 9, 5, 1, 4, 3, 8]
board = [None] * 9

# Fonction pour afficher le plateau de jeu avec des lignes horizontales et verticales
def print_board():
    horizontal_line = "+---+---+---+"
    vertical_line = "|"
    print(horizontal_line)
    for i in range(3):
        row = " "
        for j in range(3):
            index = i * 3 + j
            value = board[index] if board[index] else index
            row += f" {value} {vertical_line}"
        print(row)
        print(horizontal_line)

# Vérifier si un joueur a gagné
def did_player_win(score):
    for i in range(len(score)):
        for j in range(i + 1, len(score)):
            for k in range(j + 1, len(score)):
                if score[i] + score[j] + score[k] == 15:
                    return True
    return False

# Obtenir le score du gagnant
def get_winner_score():
    if did_player_win(score_human):
        return -1
    elif did_player_win(score_ai):
        return 1
    elif not can_play():
        return 0
    return None

# Algorithme Minimax avec élagage Alpha-Beta
def minimax(copy_board, is_maximizing, alpha, beta):
    winner_score = get_winner_score()
    if winner_score is not None:
        return winner_score

    if is_maximizing:
        best_score = -math.inf
        for index in range(9):
            if copy_board[index] is None:
                copy_board[index] = ai
                score_ai.append(magic_board[index])
                best_score = max(best_score, minimax(copy_board, False, alpha, beta))
                copy_board[index] = None
                score_ai.pop()
                alpha = max(alpha, best_score)
                if beta <= alpha:
                    break
        return best_score
    else:
        best_score = math.inf
        for index in range(9):
            if copy_board[index] is None:
                copy_board[index] = human
                score_human.append(magic_board[index])
                best_score = min(best_score, minimax(copy_board, True, alpha, beta))
                copy_board[index] = None
                score_human.pop()
                beta = min(beta, best_score)
                if beta <= alpha:
                    break
        return best_score

# Trouver le meilleur mouvement
def find_best_move():
    best_score = -math.inf
    move = None

    for index in range(9):
        if board[index] is None:
            board[index] = ai
            score = minimax(board, False, -math.inf, math.inf)
            board[index] = None

            if score > best_score:
                best_score = score
                move = index

    return move

# Vérifier s'il reste des mouvements possibles
def can_play():
    return None in board

# Valider une commande de mouvement
def is_command_valid(command):
    return command.isdigit() and int(command) in range(9)

# Boucle principale pour exécuter le jeu
def game_loop():
    print(f"{human} à ton tour")
    while True:
        command = input("Entrez votre mouvement (0-8): ")
        if not is_command_valid(command):
            print("Erreur: commande invalide")
        else:
            command = int(command)
            if board[command] is not None:
                print("Erreur: un joueur occupe déjà cette place")
            else:
                board[command] = human
                score_human.append(magic_board[command])
                if did_player_win(score_human):
                    print_board()
                    print(f"{human} a gagné")
                    break

                move = find_best_move()
                board[move] = ai
                score_ai.append(magic_board[move])
                if did_player_win(score_ai):
                    print_board()
                    print(f"{ai} a gagné")
                    break

                if not can_play():
                    print_board()
                    print("Égalité !")
                    break

                print_board()
                print(f"{human} à ton tour")

if __name__ == "__main__":
    move = find_best_move()
    board[move] = ai
    score_ai.append(magic_board[move])
    print_board()
    game_loop()
```
